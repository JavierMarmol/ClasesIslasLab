$$
f'(x) = \frac{f(x+h)-f(x)}{h}
$$

```python
def derivador(func,x,h):
    y = (func(x+h) - func(x))/h
    return y
```

```python
plot(x,func(x))
plot(x,derivador(func,x,0.5))
plot(x,2*x)
```

```python
def mi_ec_dif(y,x):
    return y
```

$$
\int_{x_i}^{x_i + h} y'(x) \simeq y_{i} + h y(x_i)
$$

```python
def euler(func,p_ini,h,x_f):
    pasos = arange(p_ini[0],x_f+h,h)
    y = zeros(len(pasos))
    y[0] = p_ini[1]
    for i in range(len(pasos)-1):
        y[i+1] = y[i] + h*func(y[i],pasos[i])
    solucion = [pasos,y]
    return solucion
```

```python
sol = euler(mi_ec_dif,[0,1],0.01,20)
plot(sol[0],sol[1])
plot(sol[0],exp(sol[0]),'red')
xlim(0,20)
```

$$
y'(x) = -xy(x)
$$

$$
y(x) = Ae^{\frac{-x^2}{2}}
$$

```python
def ec_2(y,x):
    return -x*y
```

```python
sol = euler(ec_2,[0,1],0.5,10)
plot(sol[0],sol[1],sol[0],exp((-sol[0]**2)/2))
```
